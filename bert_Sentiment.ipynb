```python
tokenizer = AutoTokenizer.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
model=AutoModelForSequenceClassification.from_pretrained('nlptown/bert-base-multilingual-uncased-sentiment')
```

```python
tokens=tokenizer.encode('i tried so hard yet,why?',return_tensors='pt')
```

```python
r=requests.get('https://magicpin.in/Mumbai/Powai/Restaurant/Hitchki/store/9370c/reviews/')
soup=BeautifulSoup(r.text,'html.parser')
regex=re.compile("review")
results=soup.find_all('p',{'class':regex})
reviews=[result.text for result in results ]
reviews
```

```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame(np.array(reviews),columns=['reviews'])
```

```python
df.tail()
```

```python
def sentiment(reviews):
  tokens=tokenizer.encode(reviews,return_tensors='pt')
  result=model(tokens)
  return int (torch.argmax((result.logits)))+1
```

```python
sentiment(df['reviews'].iloc[0])
```

```python
df['reviews'].iloc[0]
```

```python
df['sentiment']=df['reviews'].apply(lambda x : sentiment(x[:512]))
```
